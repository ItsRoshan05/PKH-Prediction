```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Memuat data dari file CSV
file_path = '../data/datajamal.csv'
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama data untuk memastikan formatnya
df.head()
```

```python
label_distribution = df['SETATUS_PENERIMA'].value_counts()
print(label_distribution)
```

```python
# Define features and target
X = df[['JUMLAH_KELUARGA', 'PENGHASILAN', 'PENDIDIKAN_TERTINGGI', 'SETATUS_RUMAH', 'PEKERJAAN', 'KONDISI_KESEHATAN']]
y = df['SETATUS_PENERIMA']
```

```python
# Initialize LabelEncoders for each categorical feature
label_encoders = {}
for column in X.columns:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

# Encode the target variable
target_le = LabelEncoder()
y = target_le.fit_transform(y)
label_encoders['SETATUS_PENERIMA'] = target_le

# Split data (training dan test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Inisialisasi model Naive Bayes
model_nb = GaussianNB()

# Latih model Naive Bayes
model_nb.fit(X_train, y_train)

# Prediksi dengan Naive Bayes
y_pred_nb = model_nb.predict(X_test)

# Evaluasi model Naive Bayes
accuracy_nb = accuracy_score(y_test, y_pred_nb)
conf_matrix_nb = confusion_matrix(y_test, y_pred_nb)
class_report_nb = classification_report(y_test, y_pred_nb)

print(f'Accuracy of Naive Bayes: {accuracy_nb}')
print('Confusion Matrix of Naive Bayes:')
print(conf_matrix_nb)
print('Classification Report of Naive Bayes:')
print(class_report_nb)

# Inisialisasi model C4.5 (DecisionTreeClassifier dengan criterion='entropy')
model_c45 = DecisionTreeClassifier(criterion='entropy', random_state=10)

# Latih model C4.5
model_c45.fit(X_train, y_train)

# Prediksi dengan C4.5
y_pred_c45 = model_c45.predict(X_test)

# Evaluasi model C4.5
accuracy_c45 = accuracy_score(y_test, y_pred_c45)
conf_matrix_c45 = confusion_matrix(y_test, y_pred_c45)
class_report_c45 = classification_report(y_test, y_pred_c45)

print(f'Accuracy of C4.5 (DecisionTreeClassifier): {accuracy_c45}')
print('Confusion Matrix of C4.5:')
print(conf_matrix_c45)
print('Classification Report of C4.5:')
print(class_report_c45)
```

```python
# Data baru untuk prediksi
data_baru = {
    'JUMLAHKELUARGA': ['Keluarga Kecil'],
    'PENGHASILAN': ['Sedang'],
    'PENDIDIKAN TERTINGGI': ['SMP'],
    'SETATUS RUMAH': ['Sedang'],
    'PEKERJAAN': ['lain-lain'],
    'KONDISI KESEHATAN': ['Cukup']
}

# Membuat DataFrame dari data baru
new_df = pd.DataFrame(data_baru)

# Encode fitur baru dengan encoder yang sama yang digunakan pada data pelatihan
for column in new_df.columns:
    if column in label_encoders:
        le = label_encoders[column]
        new_df[column] = le.transform(new_df[column])

# Melakukan prediksi dengan Naive Bayes
prediction_nb = model_nb.predict(new_df)
predicted_status_nb = label_encoders['SETATUS_PENERIMA'].inverse_transform(prediction_nb)

# Melakukan prediksi dengan C4.5 (DecisionTreeClassifier)
prediction_c45 = model_c45.predict(new_df)
predicted_status_c45 = label_encoders['SETATUS_PENERIMA'].inverse_transform(prediction_c45)

# Tampilkan hasil prediksi
print(f'Prediksi dengan Naive Bayes: {predicted_status_nb[0]}')
print(f'Prediksi dengan C4.5 (DecisionTreeClassifier): {predicted_status_c45[0]}')


```

```python
import joblib

# Simpan model Naive Bayes
joblib.dump(model_nb, '../model/model_nb.joblib')

# Simpan model C4.5 (DecisionTreeClassifier)
joblib.dump(model_c45, '../model/model_c45.joblib')


joblib.dump(label_encoders, '../model/label_encoders.joblib')
```
