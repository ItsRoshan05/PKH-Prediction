```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Memuat data dari file CSV
file_path = 'data/datajamal.csv'
df = pd.read_csv(file_path)

# Menyusun data (misalnya, menghapus kolom yang tidak relevan)
df = df.drop(columns=['NO', 'NAMA', 'NIK', 'ALAMAT'])  # Sesuaikan dengan kolom yang relevan

# Hitung distribusi label
label_distribution = df['SETATUS PENERIMA'].value_counts()
print(label_distribution)

# Set ukuran gambar
plt.figure(figsize=(10, 6))

# Plot distribusi label menggunakan seaborn
sns.countplot(x='SETATUS PENERIMA', data=df, palette='viridis')

# Tambahkan judul dan label
plt.title('Distribusi Label SETATUS PENERIMA', fontsize=16)
plt.xlabel('SETATUS PENERIMA', fontsize=14)
plt.ylabel('Jumlah', fontsize=14)

# Tampilkan plot
plt.show()
```
